```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.arange(-15,35,.5)
```

```python
x
```

```python
x=x.reshape(-1,1) #column 1uta banauni, row -1 means jati ota row bancha teti ota banauni
```

```python
m=0.7
c=3.78
y= m*x+c
```

```python
y[:10]
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y,marker=".")
plt.grid()
plt.gca().set_aspect("equal")
```

```python
model1=LinearRegression()
```

```python
model1.fit(x,y)
```

```python
model1.coef_
```

```python
model1.intercept_
```

```python
model1.predict(36)
```

```python
#noise
x.shape
```

```python
np.random.seed(10)
noise=np.random.randn(x.shape[0],1)*2
```

```python
noise[:10]
```

```python
plt.plot(noise)
```

```python
noise_y=y+noise
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x,noise_y,marker=".")
plt.grid()
plt.gca().set_aspect("equal")
```

```python
model2=LinearRegression()
```

```python
model2.fit(x,noise_y)
```

```python
noise_y_pred=model2.predict(x)
plt.figure(figsize=(10,10))
plt.scatter(x,noise_y,marker=".",label="Actual")
plt.plot(x,noise_y_pred,color='r',label="Predicted")
plt.grid()
plt.legend()
plt.xlabel("X color")
plt.ylabel("Y color")
plt.gca().set_aspect("equal")
```
